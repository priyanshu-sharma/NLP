**NLP CHINKING**

```python
import nltk
from nltk.corpus import state_union
from nltk.tokenize import PunktSentenceTokenizer
```

**LOADING TRAINING CORPORA**

```python
train = state_union.raw("2005-GWBush.txt")   
```

```python
print(train)
```

**LOADING TESTING CORPORA**

```python
test = state_union.raw("2006-GWBush.txt")
```

```python
print(test)
```

```python
custom_sent_tokenizer = PunktSentenceTokenizer(train)
```

```python
print(custom_sent_tokenizer)
```

```python
tokenized = custom_sent_tokenizer.tokenize(test)
```

**SENTENCE TOKENIZING THE COMPLETE CORPORA**

```python
print(tokenized)
```

**TOKENIZING EACH SENTENCE INTO WORD TOKENIZATION FORM THEN GO FOR THE POS TAGGING AND THEN FINALLY GOING FOR CHINKING**

```python
def process():
    try:
        for i in tokenized:
            words = nltk.word_tokenize(i)  ##sentence tokenization to word tokenization
            tagged = nltk.pos_tag(words)   ##after word tokenization we go for the pos tagging
            chinkGram = r"""Chunk: {<.*>+}
                                    }<VB.?|IN|DT|TO>+{"""
            
            chinkParser = nltk.RegexpParser(chinkGram)
            chinked = chinkParser.parse(tagged)
            #chinked.draw()
            print(chinked)
            
    except Exception as e:
        print(str(e))
```

```python
process()
```
